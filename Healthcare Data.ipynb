##### **Univariate Analysis**

```python
#Creating Subplots
plt.figure(figsize=(12, 8))

#First Subplot-Age Distribution
plt.subplot(3,2,1)
sns.histplot(data=data,x="Age",bins=16,color="deepskyblue",edgecolor='black')
plt.title("Age Wise Distribution of Patients",fontweight='bold')
plt.xlabel('Age',size=8)
plt.ylabel('Count of patients',size=8)
# plt.xticks(rotation=45,ha='right')

#Second Subplot for Gender Distribution
plt.subplot(3,2,2)
color=['deepskyblue','lightskyblue']
sns.countplot(data=data,x="Gender",palette=color,hue="Gender",edgecolor='black',legend=False)
plt.title("Gender wise Distribution",fontweight='bold')
plt.xlabel("Gender",size=8)
plt.ylabel("Count of Patients", size=8)
# plt.xticks(rotation=45,ha='right')

# Third Subplot-Medical Condition Destribution
plt.subplot(3,2,3)
sns.countplot(data=data,x="Medical Condition",color='deepskyblue',edgecolor='black')
plt.title("Medical Condition wise Destribution",fontweight='bold')
plt.xlabel("Medical condition",size=8)
plt.ylabel("Count of Patients",size=8)
plt.xticks(rotation=45,ha='right')

#Fourth Subplot-Insurance wise Destribution
plt.subplot(3,2,4)
sns.countplot(data=data,x='Insurance Provider',color='deepskyblue',edgecolor='black')
plt.title("Insurance Provider wise Destribution",fontweight='bold')
plt.xlabel('Insurance Providers',size=8)
plt.ylabel('Count of Patients',size=8)
plt.xticks(rotation=45,ha='right')

# Fifth Subplot-Blood Type Destribution
plt.subplot(3,2,5)
sns.countplot(data=data,x='Blood Type',color='deepskyblue',edgecolor='black')
plt.title("Blood Type wise Destribution",fontweight='bold')
plt.xlabel('Blood Type',size=8)
plt.ylabel('Count of Patients',size=8)
# plt.xticks(rotation=45,ha='right')

# Sixth Subplot-Test Result Destribution
plt.subplot(3,2,6)
sns.countplot(data=data, x='Test Results', color='deepskyblue', edgecolor='black')
plt.title('Test Result wise Destribution', fontweight='bold')
plt.xlabel('Test Results', size=8)
plt.ylabel('Count of Patients', size=8)
plt.xticks(rotation=45,ha='right')

plt.tight_layout()
plt.show()
```

#### **Multivariate Analysis**

##### **Condition × Admission Type × Test Result**

```python
# Set professional plotting style
sns.set_style("whitegrid")
plt.rcParams.update({
    'font.size': 10,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'figure.titlesize': 16,
    'axes.spines.top': False,
    'axes.spines.right': False
})

# Build 3-way cross-tab: Condition × Admission Type → % of Test Results
cross_tab = pd.crosstab(
    index=[data['Medical Condition'], data['Admission Type']],
    columns=data['Test Results'],
    normalize='index'  # Normalize by row (i.e., % within each condition+admission combo)
) * 100  # Convert to percentage

# Reshape for heatmap compatibility
heatmap_data = cross_tab.unstack(level=-1).fillna(0)  # Shape: (Admission Type × Condition) × Test Results

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_data.T,
    annot=True,
    fmt=".1f",
    cmap="YlGnBu",
    cbar_kws={'label': 'Percentage of Test Results (%)'}
)
plt.title("Test Results Distribution by Medical Condition & Admission Type", fontweight='bold')
plt.xlabel("")
plt.ylabel("Test Results")
plt.tight_layout()
plt.show()
```

##### **Distribution of Medical Conditions across Blood Types**

```python
# Create a pivot table: count of patients for each Blood Type × Medical Condition
heatmap_data = data.pivot_table(
    index='Medical Condition',
    columns='Blood Type',
    values='Name',  
    aggfunc='count'
).fillna(0)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data,
    annot=True,           
    fmt='g',              
    cmap='Blues',         
    cbar_kws={'label': 'Number of Patients'}
)

# Labels and title
plt.title("Medical Conditions by Blood Type", fontsize=14, fontweight='bold')
plt.xlabel("Blood Type")
plt.ylabel("Medical Condition")
plt.tight_layout()
plt.show()
```
